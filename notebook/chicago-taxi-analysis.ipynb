```python
#import required libraries for analysis 
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
#read csv files
df_company_trips = pd.read_csv('C:/Users/JGCch/OneDrive/Desktop/chicago-taxi-analysis/data/moved_project_sql_result_01.csv') 
df_dorpoff_trips = pd.read_csv('C:/Users/JGCch/OneDrive/Desktop/chicago-taxi-analysis/data/moved_project_sql_result_04.csv')
```

```python
#validate type of data and null values
df_company_trips.info()
print()
df_dorpoff_trips.info()
```

```python
#ordenate ascending way second dataset
sort_dropoff_trips = df_dorpoff_trips.sort_values(by='average_trips',ascending=False)
```

```python
#validate first lines
sort_dropoff_trips.head()
```

```python
#graph to compare first dataset. Amount of trips of each company
top_company_trips = df_company_trips.head(10)
top_company_trips.plot(title = 'Top companies vs amount of trips',
                       kind='bar',
                       xlabel = 'Company Name',
                       x = 'company_name'
                      )
plt.grid(True)
plt.show()
```

```python
#top 10 neighborhoods where trips are finished
sort_dropoff_trips = sort_dropoff_trips.head(10)
sort_dropoff_trips.plot(title = 'Top 10 neighborhood where trips are finished',
                       kind='bar',
                       xlabel = 'Dropoff location name',
                       x = 'dropoff_location_name',
                       rot = 45
                      )
plt.grid(True)
plt.show()
sort_dropoff_trips.head()
```
